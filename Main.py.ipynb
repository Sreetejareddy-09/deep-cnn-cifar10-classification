PREPROCESSING

```python
from tensorflow.keras.preprocessing import image
```

```python
filename="/content/image classification-1.jpeg"
img=image.load_img(filename,target_size=(224,224))
plt.imshow(img)
```

```python
#convert image to array and preprocessing the image
from typing_extensions import final
resized_img=image.img_to_array(img)
final_img=np.expand_dims(resized_img,axis=0)
final_img=tf.keras.applications.mobilenet_v2.preprocess_input(final_img)
```

```python
final_img.shape
```

```python
#prediction making
from tensorflow.keras.applications import MobileNetV2

mobile = MobileNetV2(weights='imagenet')

# Make predictions using the preprocessed image (final_img)
predictions = mobile.predict(final_img)
```

Predicting using V1 of mobilenet

```python
mobile=tf.keras.applications.mobilenet.MobileNet()
```

```python
predictions=mobile.predict(final_img)
```

```python
from tensorflow.keras.applications.imagenet_utils import decode_predictions

results = decode_predictions(predictions)
print(results)
```

Predicting using V2 of mobienet

```python
mobile= tf.keras.applications.mobilenet_v2.MobileNetV2()
```

```python
predictions=mobile.predict(final_img)
```

```python
from tensorflow.keras.applications.imagenet_utils import decode_predictions

mobile= tf.keras.applications.mobilenet_v2.MobileNetV2()
predictions=mobile.predict(final_img)
results=decode_predictions(predictions) # Use the imported decode_predictions function directly
print(results) #V2
```
